# Scatterplots

```python
def make_scatter_dashboard(series):
    
    x = df['Date']
    y = series
    rolling = 7
    y2 = round(y.rolling(rolling).mean(), 2)   
    title= f"{y.name} Over Time"
    
    app = jupyter_dash.JupyterDash(__name__, external_stylesheets=external_stylesheets, title=f"{title} Dashboard")
    
    fig = go.Figure(data=[go.Scatter(x=x, y=y, name=f'{y.name}', mode='markers'),
                          go.Scatter(x=x, y=y2, name=f"{rolling} Day Rolling Average {y.name}", marker_color='blue')])

    fig.update_layout(title=dict(text=title, 
                                 yanchor="top",
                                 y=.85,
                                 xanchor="left",
                                 x=.075),
                      legend=dict(orientation="h",
                                  yanchor="bottom",
                                  y=1.02,
                                  xanchor="right",
                                  x=1),
                      width=1000)

    fig.update_xaxes(title_text=f"{x.name}")
    fig.update_yaxes(title_text=f"{y.name}")

    app.layout = html.Div([
        dcc.Graph(
            figure=fig
        )
    ])
    
    fig.write_image(f"data/images/fitbit/scatterplots/{title.replace('/', '_')}.png")

    if __name__ == '__main__':
        app.run_server(mode='external', port=8020, debug=True)
```

# Histograms

```python
def make_hist_dashboard_with_bins(series, 
                                  #value_range, 
                                  #bins, 
                                  tickangle):
    title = f"{series.name} Histogram"
    app = jupyter_dash.JupyterDash(__name__, external_stylesheets=external_stylesheets, title=f"{title} Dashboard")
    
    fig = go.Figure(data=[go.Histogram(x=series)])
    
    fig.update_layout(title=dict(text=title, 
                                 yanchor="top",
                                 y=.85,
                                 xanchor="left",
                                 x=.075),
                      legend=dict(orientation="h",
                                  yanchor="bottom",
                                  y=1.02,
                                  xanchor="right",
                                  x=1),
                     bargap=0.1, 
                     width=1000)

    fig.update_traces(xbins_start=min(series), 
                      #xbins_size=bins, 
                      selector=dict(type='histogram'))
    fig.update_xaxes(tickmode='array', ticks="outside", tickson="boundaries", 
                     #ticktext=np.arange(value_range[0], value_range[1], 1), 
                     #tickvals=np.arange(value_range[0], value_range[1], 1)+.5, 
                     #range=value_range, 
                     tickangle=tickangle)
    fig.update_xaxes(title_text=f"{series.name} Level")
    fig.update_yaxes(title_text=f"{series.name} Count")
    
    app.layout = html.Div([
       dcc.Graph(
           figure=fig
       )
    ])

    title = title.replace('/', '_').replace('<br>', '')
    fig.write_image(f"data/images/fitbit/histograms/{title}.png")

    if __name__ == '__main__':
       app.run_server(mode='external', debug=True, port=8071)
```

# Github Heatmaps

```python
colors='#636EFA'
```

```python
# https://python.plainenglish.io/interactive-calendar-heatmaps-with-plotly-the-easieast-way-youll-find-5fc322125db7
# https://community.plotly.com/t/colored-calendar-heatmap-in-dash/10907/21
# https://github.com/brunorosilva/plotly-calplot


def make_github_heatmap_dashboard(df=df, 
                            series=None,
                            colors=colors, **kwargs):

    df['Date'] = pd.to_datetime(df['Date'])

    title = f'{series} Timeline Heatmap'

    fig = calplot(
             df,
             x='Date',
             y=series,
             dark_theme=True,
             years_title=True,
             showscale=True,
             colorscale=[(0.00, "white"), (1.00, colors)],
             title=title,
             month_lines_width=3, 
             month_lines_color="#444"
    )

    fig.update_layout(**kwargs)
    title = title.replace('/', '_')
    fig.write_image(f"data/images/fitbit/github_heatmaps/{title}.png")

    app = jupyter_dash.JupyterDash(__name__,
                                    external_stylesheets=external_stylesheets,
                                    title=f"{title}")
                  
    app.layout = html.Div([
        dcc.Graph(
            figure=fig,
            
            #https://plotly.com/python/setting-graph-size/
            #https://stackoverflow.com/questions/46287189/how-can-i-change-the-size-of-my-dash-graph
            style={'height': '98vh',
                   'width': '98vw',
                  },
            
    #         style={'height': '950px',
    #                'width': '1600px',
    #               },
        ),
    ])
    
    if __name__ == '__main__':
        app.run_server(mode='external', port=8110, debug=False)
```

```python
for col in df.columns[1:]:
    print(col)
    make_github_heatmap_dashboard(df=df, colors=colors, series=col, width=1600, height=1600, 
                                  title=dict(yanchor="top",y=.99, xanchor="left", x=.05))
```
