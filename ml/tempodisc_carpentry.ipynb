```python
# https://stackoverflow.com/questions/51656065/pandas-resampling-typeerror-only-valid-with-datetimeindex-timedeltaindex-or-p
# https://stackoverflow.com/questions/54592536/find-gaps-in-pandas-time-series-dataframe-sampled-at-1-minute-intervals-and-fill
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.resample.html
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.asfreq.html#pandas.DataFrame.asfreq
df = df.set_index('Date')
df = df.resample('T', convention='start').asfreq() # 'T' = minutes, 'D' = days, 's' = seconds
df = df.reset_index()
df
```

# Scatterplots

```python
def make_scatter_dashboard(series):
    
    x = df['Date']
    y = series
    rolling = 7
    y2 = round(y.rolling(rolling).mean(), 2)   
    title= f"{y.name} Over Time"
    
    app = jupyter_dash.JupyterDash(__name__, external_stylesheets=external_stylesheets, title=f"{title} Dashboard")
    
    fig = go.Figure(data=[go.Scatter(x=x, y=y, name=f'{y.name}', mode='markers', #marker_color='red'
                                    ),
                          go.Scatter(x=x, y=y2, name=f"{rolling} Day Rolling Average {y.name}", marker_color='blue')])

    fig.update_layout(title=dict(text=title, 
                                 yanchor="top",
                                 y=.85,
                                 xanchor="left",
                                 x=.075),
                      legend=dict(orientation="h",
                                  yanchor="bottom",
                                  y=1.02,
                                  xanchor="right",
                                  x=1),
                      width=1000)

    fig.update_xaxes(title_text=f"{x.name}")
    fig.update_yaxes(title_text=f"{y.name}")

    app.layout = html.Div([
        dcc.Graph(
            figure=fig
        )
    ])
    fig.write_image(f"data/images/tempodisc/{title.replace('/', '_')}.png")

    if __name__ == '__main__':
        app.run_server(mode='external', port=8030, debug=False)
```

```python
df = df.rename(columns={'temperature': 'Ambient Temperature', 
                        'humidity':'Ambient Humidity', 
                        'dewpoint':'Ambient Dewpoint', 
                        'pressure':'Ambient Pressure'})
df
```

```python
make_scatter_dashboard(df['Ambient Temperature'])
```

```python
make_scatter_dashboard(df['Ambient Humidity'])
```

```python
make_scatter_dashboard(df['Ambient Dewpoint'])
```

```python
make_scatter_dashboard(df['Ambient Pressure'])
```
