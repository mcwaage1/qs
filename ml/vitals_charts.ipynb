# Scatterplots

```python
def make_scatter_dashboard(series):
    
    x = df['Date']
    y = series
    rolling = 7
    y2 = round(y.rolling(rolling).mean(), 2)   
    title= f"{y.name} Over Time"
    
    app = jupyter_dash.JupyterDash(__name__, external_stylesheets=external_stylesheets, title=f"{title} Dashboard")
    
    fig = go.Figure(data=[go.Scatter(x=x, y=y, name=f'{y.name}', mode='markers', #marker_color='red'
                                    ),
                          go.Scatter(x=x, y=y2, name=f"{rolling} Day Rolling Average {y.name}", marker_color='blue')])

    fig.update_layout(title=dict(text=title, 
                                 yanchor="top",
                                 y=.85,
                                 xanchor="left",
                                 x=.075),
                      legend=dict(orientation="h",
                                  yanchor="bottom",
                                  y=1.02,
                                  xanchor="right",
                                  x=1),
                      width=1000)

    fig.update_xaxes(title_text=f"{x.name}")
    fig.update_yaxes(title_text=f"{y.name}")

    app.layout = html.Div([
        dcc.Graph(
            figure=fig
        )
    ])
    fig.write_image(f"data/images/vitals/scatterplots/{title.replace('/', '_')}.png")

    if __name__ == '__main__':
        app.run_server(mode='external', port=8030, debug=False)
```

```python
# https://plotly.com/python/multiple-axes/
def make_dashboard_with_secondary_axis(series1, series2, rolling, value_range1, value_range2):
    
    x = df['Date']
    y = series1
    y2= series2
    y_rolling = round(y.rolling(rolling).mean(), 2)
    y2_rolling = round(y2.rolling(rolling).mean(), 2)
    title = f"{y.name} and {y2.name} Over Time"
    
    app = jupyter_dash.JupyterDash(__name__, external_stylesheets=external_stylesheets, title=f"{title} Dashboard")
    
    app.layout = dcc.Loading(
        html.Div([
            dcc.Graph(id="graph"),
            html.P("Axes:"),
            dcc.RadioItems(
                id='radio',
                value='Secondary',
                options=[{'label': x, 'value': x}
                         for x in ['Primary', 'Secondary']]
            )
        ]))

    @app.callback(
        Output("graph", "figure"), 
        [Input("radio", "value")])

    def display_(radio_value):

        # Create figure with secondary y-axis
        fig = make_subplots(specs=[[{"secondary_y": True}]])
        
        # y
        fig.add_trace(
            go.Scatter(x=x, 
                       y=y, 
                       name=f'{y.name}', 
                       marker_color='blue', 
                       mode='markers', 
                       #visible='legendonly'
                      ), 
            secondary_y=False)
        # y rolling
        fig.add_trace(
            go.Scatter(x=x, 
                       y=y_rolling, 
                       name=f"{rolling} Day Rolling Average {y.name}", 
                       marker_color='blue',
                       yaxis='y2'), 
            secondary_y=False)

        # y2
        fig.add_trace(
            go.Scatter(x=x, 
                       y=y2, 
                       name=f'{y2.name}', 
                       marker_color='red', 
                       mode='markers', 
                       #visible='legendonly',
                       yaxis='y3'), 
            secondary_y=radio_value == 'Secondary')
        # y2 rolling
        fig.add_trace(
            go.Scatter(x=x, 
                       y=y2_rolling, 
                       name=f"{rolling} Day Rolling Average {y2.name}", 
                       marker_color='red',
                       yaxis='y4'), 
            secondary_y=radio_value == 'Secondary')
        
        # Set x-axis title
        fig.update_xaxes(title_text=f"{x.name}")

        # Set y-axes titles
        fig.update_yaxes(
            title_text=f"{y.name}",
            range=value_range1,
            secondary_y=False,
            autorange=False)
        fig.update_yaxes(
            title_text=f"{y2.name}",
            range=value_range2,
            secondary_y=True,
            showgrid=False,
            autorange=False)
        
        fig.update_layout(title=dict(text=title, 
                                 yanchor="top",
                                 y=.95,
                                 xanchor="left",
                                 x=.075),
                          # legend=dict(orientation="h",
                          #         yanchor="bottom",
                          #         y=1.02,
                          #         xanchor="right",
                          #         x=1),
                         width=1000,
                         height=1000
                         )
                          
        return fig

    fig.write_image(f"data/images/vitals/scatterplots/{title.replace('/', '_')}.png")
    
    if __name__ == '__main__':
        app.run_server(mode='external', debug=True, port=8061)
```

# Histograms

```python
def make_hist_dashboard(series):
    title = f"{series.name} Histogram"
    app = jupyter_dash.JupyterDash(__name__, external_stylesheets=external_stylesheets)
    
    fig = go.Figure(data=[go.Histogram(x=series)])
    fig.update_xaxes(title_text=f"{series.name} Level")
    fig.update_yaxes(title_text=f"{series.name} Count")
    fig.update_layout(title=dict(text=title, 
                                 yanchor="top",
                                 y=.85,
                                 xanchor="left",
                                 x=.075),
                      legend=dict(orientation="h",
                                  yanchor="bottom",
                                  y=1.02,
                                  xanchor="right",
                                  x=1),
                     bargap=0.1)

    app.layout = html.Div([
        dcc.Graph(
            figure=fig
        )
    ])
    title = title.replace('/', '_').replace('<br>', '')
    fig.write_image(f"data/images/vitals/histograms/{title}.png")

    if __name__ == '__main__':
        app.run_server(mode='external', debug=True, port=8081)
```

```python
def make_hist_dashboard_with_bins(series, value_range, bins, tickangle):
    title = f"{series.name} Histogram"
    app = jupyter_dash.JupyterDash(__name__, external_stylesheets=external_stylesheets, title=f"{title} Dashboard")
    
    fig = go.Figure(data=[go.Histogram(x=series)])
    
    fig.update_layout(title=dict(text=title, 
                                 yanchor="top",
                                 y=.85,
                                 xanchor="left",
                                 x=.075),
                      legend=dict(orientation="h",
                                  yanchor="bottom",
                                  y=1.02,
                                  xanchor="right",
                                  x=1),
                     bargap=0.1)

    fig.update_traces(xbins_start=min(series), xbins_size=bins, selector=dict(type='histogram'))
    fig.update_xaxes(tickmode='array', ticks="outside", tickson="boundaries", 
                     ticktext=np.arange(value_range[0], value_range[1], 1), 
                     tickvals=np.arange(value_range[0], value_range[1], 1), 
                     range=value_range, 
                     tickangle=tickangle)
    fig.update_xaxes(title_text=f"{series.name} Level")
    fig.update_yaxes(title_text=f"{series.name} Count")
    
    app.layout = html.Div([
       dcc.Graph(
           figure=fig
       )
    ])

    title = title.replace('/', '_').replace('<br>', '')
    fig.write_image(f"data/images/vitals/histograms/{title}.png")

    if __name__ == '__main__':
       app.run_server(mode='external', debug=False, port=8072)
```

```python
make_hist_dashboard_with_bins(df['Temperature'], [95.5, 98], .1, 0)
```

# Github Heatmaps

```python
colors='#636EFA' 
# colors='#EF553B'
```

```python
# https://python.plainenglish.io/interactive-calendar-heatmaps-with-plotly-the-easieast-way-youll-find-5fc322125db7
# https://community.plotly.com/t/colored-calendar-heatmap-in-dash/10907/21
# https://github.com/brunorosilva/plotly-calplot


def make_github_heatmap_dashboard(df=df, 
                            series=None,
                            colors=colors, **kwargs):

    df['Date'] = pd.to_datetime(df['Date'])

    title = f'{series} Timeline Heatmap'

    fig = calplot(
             df,
             x='Date',
             y=series,
             dark_theme=True,
             years_title=True,
             showscale=True,
             colorscale=[(0.00, "white"), (1.00, colors)],
             title=title,
             month_lines_width=3, 
             month_lines_color="#444"
    )

    fig.update_layout(**kwargs)
    title = title.replace('/', '_')
    fig.write_image(f"data/images/vitals/github_heatmaps/{title}.png")

    app = jupyter_dash.JupyterDash(__name__,
                                    external_stylesheets=external_stylesheets,
                                    title=f"{title}")
                  
    app.layout = html.Div([
        dcc.Graph(
            figure=fig,
            
            #https://plotly.com/python/setting-graph-size/
            #https://stackoverflow.com/questions/46287189/how-can-i-change-the-size-of-my-dash-graph
            style={'height': '98vh',
                   'width': '98vw',
                  },
            
    #         style={'height': '950px',
    #                'width': '1600px',
    #               },
        ),
    ])
    
    if __name__ == '__main__':
        app.run_server(mode='external', port=8111, debug=False)
```

```python
for col in df.columns[5:6]:
    print(col)
    make_github_heatmap_dashboard(df=df, colors=colors, series=col, width=1600, height=1600, 
                                  title=dict(yanchor="top",y=.99, xanchor="left", x=.05))
```
